```python
#bokeh basics
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
from bokeh.models import Range1d, PanTool, ResetTool, HoverTool, ColumnDataSource
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head()
```

# Editable variables

```python
#filename = 'un_bar.html'

page_width = 596
page_height = 160

y_baseline = 0
y_maxline = 800
numTickY = 5
```

# Bar chart code

```python
output_notebook()
```

```python
#Create new location variable for x values
xLoc = df.index.values.tolist()
xLoc = [x+.5 for x in xLoc]
#add to dataframe 
df['location'] = xLoc
```

```python
#source
source.data['xLoc'] = [x+1 for x in source.data['index']]

source = ColumnDataSource(df)
countries = df['code'].tolist()
xLoc = df.index.values.tolist()
xLoc = [x+1 for x in xLoc]


#Setup output file
#output_file(filename)

#Set up figure
f = figure(x_range = countries, tools = '')
f.toolbar.logo = None

f.plot_width = page_width
f.plot_height = page_height

f.y_range = Range1d(start = y_baseline, end = y_maxline)
f.yaxis[0].ticker.desired_num_ticks = numTickY



f.axis.minor_tick_line_color = None
f.xgrid.grid_line_color = None

f.vbar(x = 'location', width = 0.7, top = 'million_people', color="#F8B617", source = source)

show(f)
```
